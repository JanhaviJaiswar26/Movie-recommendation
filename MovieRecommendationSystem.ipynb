```python
#merge both the data set onto title

movies=movies.merge(credits,on='title')
movies
```

```python

movies =movies.drop(['crew_y','movie_id_y','cast_y'],axis=1)
movies=movies.rename(columns={'crew_x':'crew','movie_id_x':'movie_id','cast_x':'cast'})
```

```python
#extracting the required columns

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(10)
```

```python
#function for formatting the columns

import ast
def convert(text):
    a = []
    for i in ast.literal_eval(text):
        a.append(i['name'])
    return a
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# dictionary
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#function to get the 3 most cast names

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
#for getting 3 most cast

movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head(10)
```

```python
#function for extraction only the director name form the list

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(10)
```

```python
movies.sample(5)
```

```python
#function for removing the space from the names


def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(10)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(5)
```

```python
#making tags from all the columns for specification


movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(5)
```

```python
#droping unwanted columns

new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head(5)
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)

```

```python
recommend('Avatar')
#Avengers: Age of Ultron
```

```python
import pickle
pickle.dump(new,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```

```python
#testing code can be remove

import re
clean=re.compile('<.*?>')
re.sub(clean,'',df.iloc[2].review)
```

```python
#  Function for cleaning the html tags

def clean_html(text):
    clean=re.compile('<.*?>')
    return re.sub(clean,'',text)
```

```python
df['review']=df['review'].apply(clean_html)
df.head()
```

```python
#function to make all letters in lower case

def convert_lower(text):
    return text.lower()
```

```python
df['review']=df['review'].apply(convert_lower)
df.head(10)
```

```python
#function to remove special chareacters form the reviews

def remove_special(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
df['review']=df['review'].apply(remove_special)
df.head(10)
```

```python
#removing the stop words

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# shows all the stop words from that language

stopwords.words('english')
```

```python
#method 1
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
#method 2 -good
from nltk.corpus import stopwords

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))  # Use set for faster lookup
    return [word for word in text.split() if word not in stop_words]
```

```python
df['review']=df['review'].apply(remove_stopwords)
df.head(10)
```

```python
#stemming means it take all similar words as coommon name from them

import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#method 1 good
y=[]

def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
#method2

from nltk.stem import PorterStemmer

def stem_words(text):
    ps = PorterStemmer()  # Instantiate the stemmer outside the loop
    return [ps.stem(word) for word in text]
```

```python
df['review']=df['review'].apply(stem_words)
df.head(10)
```

```python
#join funcuation

def join_back(list_input):
    return " ".join(list_input)
```

```python
df['review']=df['review'].apply(join_back)
df.head(10)
```
